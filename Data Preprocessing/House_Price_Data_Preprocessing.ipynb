1. Importing the Necessary Libraries

```python
import pandas as pd
from datetime import datetime
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

2. Importing the Dataset

```python
dataset = pd.read_csv("/content/DKHousingPricesSample100k.csv")
dataset
```

```python
dataset = dataset.dropna(axis = 0)
```

```python
dataset['region'] = dataset['region'].replace({'Fyn & islands': 'Fyn_and_islands'})
```

```python
dataset['date'] = pd.to_datetime(dataset['date']) # Convert date column to datetime format
current_year = datetime.now().year # Get the current year
dataset['age'] = current_year - dataset['year_build'] # Calculate the age of the house
dataset
```

```python
dataset = dataset.drop(['date' , 'quarter' , 'house_id' , 'year_build' , 'sqm_price' , 'address' , 'city' , 'zip_code' , 'area'] , axis='columns')
```

```python
dataset
```

```python
def remove_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df
```

```python
outliers = ["purchase_price", "%_change_between_offer_and_purchase", "no_rooms" ,
                      'sqm' , "nom_interest_rate%" , "dk_ann_infl_rate%" , "yield_on_mortgage_credit_bonds%" , "age"]

for feature in outliers :
    dataset = remove_outliers(dataset, feature)
```

```python
dataset.isnull().sum()
```

```python
needs_to_be_scaled = ["purchase_price", "%_change_between_offer_and_purchase", "no_rooms" ,
                      'sqm' , "nom_interest_rate%" , "dk_ann_infl_rate%" , "yield_on_mortgage_credit_bonds%" , "age"]

for feature in needs_to_be_scaled:
    scaler = StandardScaler()
    dataset[feature] = scaler.fit_transform(dataset[[feature]])
```

```python
multi_class_feature = ['house_type' , 'sales_type' , 'region']

# Apply OneHotEncoding only once
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Encode categorical features
encoded_data = encoder.fit_transform(dataset[multi_class_feature])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(multi_class_feature))

# Drop original categorical features and concatenate the encoded data
dataset = dataset.drop(columns=multi_class_feature)
dataset = pd.concat([dataset, encoded_df], axis=1)
```

```python
# Save the processed dataset to a new CSV file
dataset.to_csv('processed_dataset.csv', index=False)
print("Processed dataset saved to 'processed_dataset.csv'")
```
