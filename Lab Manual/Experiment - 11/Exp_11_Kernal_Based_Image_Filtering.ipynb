```python
# 1. Importing the Necessary Libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageFilter
from scipy. ndimage.filters import convolve
from google.colab.patches import cv2_imshow
```

```python
# 2. Importing the Image

image_path = '/content/Image.jpg'
img = cv2.imread(image_path ,  cv2.IMREAD_GRAYSCALE)
image_float = np.array(img, dtype=np.float64)
cv2_imshow(img)
```

Now Implementing Differnet types of Kernals

1. Identity Kernel

```python
print("Kernel: Identity Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[0, 0, 0],
                   [0, 1, 0],
                   [0, 0, 0]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

2. Box Blur (Average Filter)

```python
print("Kernel: Box Blur Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = (1/9) * np.array([[1, 1, 1],
                           [1, 1, 1],
                           [1, 1, 1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

3. Gaussian Blur

```python
print("Kernel: Gaussian Blur Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = (1/16) * np.array([[1, 2, 1],
                            [2, 4, 2],
                            [1, 2, 1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

4. Sobel X (Horizontal Edge Detection)

```python
print("Kernel: Sobel X Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

5. Sobel Y (Vertical Edge Detection)

```python
print("Kernel: Sobel Y Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

6. Laplacian (Edge Detection)

```python
print("Kernel: Laplacian Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[0, -1, 0],
                   [-1, 4, -1],
                   [0, -1, 0]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

7. Sharpen

```python
print("Kernel: Sharpen Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

8. High-Pass Filter

```python
print("Kernel: High-Pass Filter Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

9. Motion Blur

```python
print("Kernel: Motion Blur Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = (1/9) * np.array([[1, 0, 0],
                           [0, 1, 0],
                           [0, 0, 1]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```

10. Emboss

```python
print("Kernel: Emboss Kernel")
image_float = np.array(img, dtype=np.float64)
kernal = np.array([[-2, -1, 0],
                   [-1, 1, 1],
                   [0, 1, 2]])
im2 = convolve(image_float, kernal)
image_bounded = np.array(np.clip(im2, 0, 255), dtype=np.uint8)
img_new = cv2.hconcat([img, image_bounded])
cv2_imshow(img_new)
```
